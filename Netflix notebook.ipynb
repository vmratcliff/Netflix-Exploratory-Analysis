```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_csv("netflix_titles.csv")
file.head()
```

```python
# dropping null value columns to avoid errors
file.dropna(inplace = True)
```

```python
# new data frame with split value columns 
netflix_df = file["listed_in"].str.split(",", n = 1, expand = True)
```

```python
# making separate genre category column from new data frame 
file["Genre Main Category"]= netflix_df[0]
```

```python
# making separate sub category column from new data frame
file["Genre Sub Category"]= netflix_df[1]
```
